#### this file aims to ensemble predictions in step3 in our report.

```python
import pandas as pds
# file paths that you want to ensemble
files = ['./submit7_res.csv', './submit5_inc.csv']
```

```python

def ensemble(list_filename):
  numeric_col = ['ALB', 'BET', 'DOL', 'LAG', 'NoF', 'OTHER', 'SHARK', 'YFT']
  n = len(list_filename)
  values = pds.read_csv(list_filename[0])[numeric_col]
  for i in range(1, n):
    values += pds.read_csv(list_filename[i])[numeric_col]
  values /= n
  final = pds.read_csv(list_filename[0])
  final[numeric_col] = values
  return final
```

```python
ensemble(files)
```

```python
# save files
ensemble(files).to_csv('./ensem.csv', index=None)
```
