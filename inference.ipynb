# ensemble part

```python
import pandas as pds
def ensemble(list_filename):
  numeric_col = ['ALB', 'BET', 'DOL', 'LAG', 'NoF', 'OTHER', 'SHARK', 'YFT']
  n = len(list_filename)
  values = pds.read_csv(list_filename[0])[numeric_col]
  for i in range(1, n):
    values += pds.read_csv(list_filename[i])[numeric_col]
  values /= n
  final = pds.read_csv(list_filename[0])
  final[numeric_col] = values
  return final
```

```python
# ensemble efficientnet b7
efffiles = ['/content/data/eff400_submit8.csv','/content/data/eff400_submit9.csv']
ensemble(efffiles).to_csv('/content/data/eff400_submit.csv', index=None)

efffiles = ['/content/data/eff400_submit.csv','/content/data/eff600_submit9.csv']
ensemble(efffiles).to_csv('/content/data/eff_submit.csv', index=None)

# ensemble resnet152
resfiles = ['/content/data/res_submit6.csv', '/content/data/res_submit7.csv']
ensemble(resfiles).to_csv('/content/data/res_submit.csv', index=None)

# ensemble regnet
regfiles = ['/content/data/reg_submit8.csv', '/content/data/reg_submit9.csv']
ensemble(regfiles).to_csv('/content/data/reg_submit.csv', index=None)

#ensemble ALL
allfiles = ['/content/data/res_submit.csv', '/content/data/inc_submit5.csv', '/content/data/reg_submit.csv', '/content/data/eff_submit.csv']
ensemble(allfiles).to_csv('/content/data/final_submission.csv', index=None)

```
